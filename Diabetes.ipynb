```python
import pandas as pd

# Load the dataset
df = pd.read_csv('diabetes_prediction_dataset.csv')

# Total number of datapoints
total_datapoints = len(df)

# Count diabetic (1) and non-diabetic (0) patients
diabetes_counts = df['diabetes'].value_counts()

print(f"Total number of datapoints: {total_datapoints}")
print("\nDiabetes status counts:")
print(diabetes_counts)

# You can also print specific counts if available
if 1 in diabetes_counts:
    print(f"Number of diabetic patients (1): {diabetes_counts[1]}")
else:
    print("No diabetic patients (1) found in the dataset.")

if 0 in diabetes_counts:
    print(f"Number of non-diabetic patients (0): {diabetes_counts[0]}")
else:
    print("No non-diabetic patients (0) found in the dataset.")
```

```python
import pandas as pd

# Separate diabetic and non-diabetic patients
diabetic_df = df[df['diabetes'] == 1]
non_diabetic_df = df[df['diabetes'] == 0]

# Get the count of diabetic patients (which will be our target count for both classes)
target_count = len(diabetic_df)

print(f"Original number of diabetic patients: {len(diabetic_df)}")
print(f"Original number of non-diabetic patients: {len(non_diabetic_df)}")
print(f"Target count for both classes after balancing: {target_count}")
```

```python
# Randomly sample non-diabetic patients to match the target count
non_diabetic_sampled = non_diabetic_df.sample(n=target_count, random_state=42) # Using random_state for reproducibility

# Concatenate the diabetic and sampled non-diabetic dataframes
balanced_df = pd.concat([diabetic_df, non_diabetic_sampled])

# Shuffle the balanced dataset to mix diabetic and non-diabetic cases
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"\nShape of the balanced dataset: {balanced_df.shape}")
print("Diabetes status counts in balanced dataset:")
print(balanced_df['diabetes'].value_counts())

# Display the first few rows of the balanced dataset
display(balanced_df.head())
```

```python
# Export the balanced dataset to a new CSV file
output_filename = 'diabetes_balanced_dataset.csv'
balanced_df.to_csv(output_filename, index=False)

print(f"\nBalanced dataset exported to '{output_filename}'")
```
